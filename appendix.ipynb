```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have your data stored in variables feature1, feature2, and cluster_labels
# feature1 and feature2 should be arrays/lists containing the values of feature 1 and feature 2 respectively
# cluster_labels should be an array/list containing the cluster labels for each data point

# Sample data (replace with your actual data)

# Generate sample data
np.random.seed(0)  # for reproducibility
num_samples = 100
feature1 = np.random.randn(num_samples)  # random values for feature 1
feature2 = np.random.randn(num_samples)  # random values for feature 2
cluster_labels = np.random.randint(0, 3, num_samples)  # random cluster labels (0, 1, or 2)

# Define colors for each cluster
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']

# Plot the scatter plot
plt.figure(figsize=(8, 6))  # Adjust size as needed
for cluster_label in set(cluster_labels):
    plt.scatter(
        [feature1[i] for i, label in enumerate(cluster_labels) if label == cluster_label],
        [feature2[i] for i, label in enumerate(cluster_labels) if label == cluster_label],
        c=colors[cluster_label],
        label=f'Cluster {cluster_label}',
        marker='x'
    )

# Add labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot with Clusters')
plt.legend()  # Show legend

# Show plot
plt.grid(True)  # Add grid
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)  # for reproducibility
num_samples = 50
treated_feature1 = np.random.randint(0, 65, num_samples)  # random values for treated feature 1
treated_feature2 = np.random.randint(0, 65, num_samples)  # random values for treated feature 2

control_feature1 = np.random.randint(30, 80, num_samples)  # random values for control feature 1
control_feature2 = np.random.randint(30, 80, num_samples)  # random values for control feature 2

# Plot the scatter plot
plt.figure(figsize=(10, 8))  # Adjust size as needed

# Plot treated units (green)
plt.scatter(
    treated_feature1,
    treated_feature2,
    c='g',
    label='Treated Units',  # Add label for treated units
    marker='x'
)

# Plot control units (blue)
plt.scatter(
    control_feature1,
    control_feature2,
    c='b',
    label='Control Units',  # Add label for control units
    marker='x'
)

# Define circle properties
circle_params = {
    'edgecolor': 'r',  # Edge color for the circle
    'fill': False,  # Do not fill the circle
    'linestyle': '--',  # Line style for the circle
}

# Define the high novelty points for each definition
high_novelty_points = {
    'Singular Comparison Novelty Score': (np.random.randint(40, 60, 5), np.random.randint(40, 60, 5)),
    'Groupwise Comparison Novelty Score': (np.random.randint(10, 30, 5), np.random.randint(10, 30, 5)),
    'Hybrid Comparison Novelty Score': (np.random.randint(70, 90, 5), np.random.randint(10, 30, 5))
}

# Plot circles around high novelty points for each definition and annotate the group
for definition, (x, y) in high_novelty_points.items():
    # Calculate the centroid of the group
    centroid_x = np.mean(x)
    centroid_y = np.mean(y)
    
    # Plot circles
    for i in range(len(x)):
        circle = plt.Circle((x[i], y[i]), radius=5, **circle_params)
        plt.gca().add_patch(circle)
    
    # Add annotation for the group
    plt.text(centroid_x, centroid_y, definition, color='r', fontsize=10, ha='center', va='center')

# Add labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot with Treated and Control Units')

# Show legend
plt.legend()

# Show plot
plt.grid(True)  # Add grid
plt.show()
```
